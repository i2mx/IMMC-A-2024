```python
import pandas as pd
```

```python
qol = pd.read_csv("data/quality of life.csv")
qol
```

```python
population = pd.read_csv("data/country populations.csv")
population
```

```python
df = pd.merge(qol, population, left_on="Country" ,right_on="Location")
df
```

```python
world_qol = sum(df['QoLI'] * df['Population2023']) / sum(df['Population2023'])
world_qol
```
